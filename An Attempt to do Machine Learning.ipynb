```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from scipy.stats import zscore
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score

def calculate_z_scores_gb2(df, column):
    z_scores = zscore(df[column])
    return z_scores

def identify_outliers_gb2(z_scores, threshold):
    outliers = abs(z_scores) > threshold
    return outliers

def remove_outliers_gb2(df, outliers):
    df_no_outliers = df[~outliers]
    return df_no_outliers

independent_vars_gb2 = ['S002VS','X003','X007','A029', 'A030', 'A032','A034', 'A035','A040', 'A041', 'A042', 'A124_02', 'A124_03', 'A124_04', 'A124_07', 'A124_08', 'A124_09']
target_y010_gb2 = 'Y010'
target_y020_gb2 = 'Y020'

z_scores_y010_gb2 = calculate_z_scores_gb2(turkey_df_cleaned, target_y010_gb2)
outliers_y010_gb2 = identify_outliers_gb2(z_scores_y010_gb2, 3)
turkey_df_no_outliers_y010_gb2 = remove_outliers_gb2(turkey_df_cleaned, outliers_y010_gb2)

z_scores_y020_gb2 = calculate_z_scores_gb2(turkey_df_cleaned, target_y020_gb2)
outliers_y020_gb2 = identify_outliers_gb2(z_scores_y020_gb2, 3)
turkey_df_no_outliers_y020_gb2 = remove_outliers_gb2(turkey_df_cleaned, outliers_y020_gb2)

X_train_y010_gb2, X_test_y010_gb2, y010_train_gb2, y010_test_gb2 = train_test_split(
    turkey_df_no_outliers_y010_gb2[independent_vars_gb2],
    turkey_df_no_outliers_y010_gb2[target_y010_gb2],
    test_size=0.2, random_state=42
)

X_train_y020_gb2, X_test_y020_gb2, y020_train_gb2, y020_test_gb2 = train_test_split(
    turkey_df_no_outliers_y020_gb2[independent_vars_gb2],
    turkey_df_no_outliers_y020_gb2[target_y020_gb2],
    test_size=0.2, random_state=42
)

scaler = StandardScaler()
X_train_y010_gb2 = scaler.fit_transform(X_train_y010_gb2)
X_test_y010_gb2 = scaler.transform(X_test_y010_gb2)

X_train_y020_gb2 = scaler.fit_transform(X_train_y020_gb2)
X_test_y020_gb2 = scaler.transform(X_test_y020_gb2)

gb_y0102 = GradientBoostingRegressor(n_estimators=500, learning_rate=0.01, max_depth=6, random_state=42)
gb_y0202 = GradientBoostingRegressor(n_estimators=500, learning_rate=0.01, max_depth=6, random_state=42)

gb_y0102.fit(X_train_y010_gb2, y010_train_gb2)
gb_y0202.fit(X_train_y020_gb2, y020_train_gb2)

y010_pred_gb2 = gb_y0102.predict(X_test_y010_gb2)
y020_pred_gb2 = gb_y0202.predict(X_test_y020_gb2)

mse_y010_gb2 = mean_squared_error(y010_test_gb2, y010_pred_gb2)
mse_y020_gb2 = mean_squared_error(y020_test_gb2, y020_pred_gb2)

r2_y010_gb2 = r2_score(y010_test_gb2, y010_pred_gb2)
r2_y020_gb2 = r2_score(y020_test_gb2, y020_pred_gb2)

cv_scores_y010_gb2 = cross_val_score(gb_y0102, X_train_y010_gb2, y010_train_gb2, cv=5, scoring='r2')
cv_scores_y020_gb2 = cross_val_score(gb_y0202, X_train_y020_gb2, y020_train_gb2, cv=5, scoring='r2')

print(f'Mean Squared Error for Y010 (Gradient Boosting): {mse_y010_gb2}')
print(f'Mean Squared Error for Y020 (Gradient Boosting): {mse_y020_gb2}')

print(f'R-squared for Y010 (Gradient Boosting): {r2_y010_gb2}')
print(f'R-squared for Y020 (Gradient Boosting): {r2_y020_gb2}')
```
