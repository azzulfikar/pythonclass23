```python
selected_columns = ['S002VS', 'COUNTRY_ALPHA', 'Y020', 'Y010', 'X003', 'X001', 'X025', 'X025R', 'X047_WVS', 'X007', 'X028','X045','X026','A173']
data = pd.read_csv("/Users/zeynepzulfikar/Downloads/WVS_TimeSeries_4_0.csv", sep=',', usecols=selected_columns)

turkey_df = data[data['COUNTRY_ALPHA'] == 'TUR']

print("Count of missing values before dropna:")
print(data.isnull().sum())

turkey_df_cleaned = turkey_df.dropna(subset=['Y010', 'Y020'])

print("\nCount of missing values after dropna:")
print(turkey_df_cleaned.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

age_group_df = turkey_df_cleaned[(turkey_df_cleaned['X003'] >= 18) & (turkey_df_cleaned['X003'] <= 29)]

not_age_group_df = turkey_df_cleaned[~((turkey_df_cleaned['X003'] >= 18) & (turkey_df_cleaned['X003'] <= 29))]

sns.lineplot(x='S002VS', y='Y010', data=age_group_df, label='Secular Values (18-29)', color='blue')
sns.lineplot(x='S002VS', y='Y020', data=age_group_df, label='Emancipative Values (18-29)', color='orange')

sns.lineplot(x='S002VS', y='Y010', data=not_age_group_df, label='Secular Values (Not 18-29)', color='green')
sns.lineplot(x='S002VS', y='Y020', data=not_age_group_df, label='Emancipative Values (Not 18-29)', color='red')

plt.title('Changes in Secular and Emancipative Values for Turkish Adults')
plt.xlabel('Wave (S002VS)')
plt.ylabel('Values')
plt.legend()
plt.show()
```
